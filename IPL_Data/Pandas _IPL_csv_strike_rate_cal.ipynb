```python
#Three condition are 
    #1) The batsman should have played more than 200 balls
    #2) The batsman should have scored in the over of 16-20
    #3) We have to get the strike rate
```

```python
#over from 16-20
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()

x=all_batsman>200

#in this we are getting the name of all hte batsman which falls under our condition of playing more than 
# 200 balls in last 16-20 overs 
death_bat=all_batsman[x].index.tolist()

all_batsman[x]
```

```python
#this line we get data only which have these bats man with above conditions
death_over[death_over['batsman'].isin(death_bat)]  
```

```python
runs=death_over[death_over['batsman'].isin(death_bat)].groupby('batsman')['batsman_runs'].sum()
balls=death_over[death_over['batsman'].isin(death_bat)].groupby('batsman')['batsman_runs'].count()
```

```python
strike=(runs/balls)*100
```

```python
strike.sort_values(ascending=False)
```
